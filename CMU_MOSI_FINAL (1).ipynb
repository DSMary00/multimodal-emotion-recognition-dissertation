```python
# Cell 1: Imports & helpers
import numpy as np
import torch
from sklearn.metrics import mean_absolute_error, accuracy_score, f1_score, confusion_matrix

# fall back to numpy if SciPy isn't installed
try:
    from scipy.stats import pearsonr
    def pearson_r(a, b):
        return float(pearsonr(np.asarray(a).ravel(), np.asarray(b).ravel())[0])
except Exception:
    def pearson_r(a, b):
        a = np.asarray(a).ravel(); b = np.asarray(b).ravel()
        if a.std() == 0 or b.std() == 0:
            return 0.0
        return float(np.corrcoef(a, b)[0, 1])
```

```python

import torch
import numpy as np


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


from unimodals.common_models import GRU, MLP, Sequential, Identity
from fusions.common_fusions import ConcatEarly

# Define encoders, fusion, and head as they were during training

encoders_eval = [Identity(), Identity(), Identity()]
fusion_eval = ConcatEarly()
head_eval = Sequential(
    GRU(409, 512, dropout=True, has_padding=False,
        batch_first=True, last_only=True),
    MLP(512, 512, 1)
)




class MultimodalModel(torch.nn.Module):
    def __init__(self, encoders, fusion, head):
        super(MultimodalModel, self).__init__()
        self.encoders = torch.nn.ModuleList(encoders)
        self.fusion = fusion
        self.head = head

    def forward(self, x):
        # x is a list: [vision, audio, text]
        encoded_features = [encoder(x[i].to(device)) for i, encoder in enumerate(self.encoders)]
        fused_features = self.fusion(encoded_features)
        output = self.head(fused_features)
        return output





# Load the state_dict into the new model instance
# Use the new state_dict filename
model_path = 'mosi_ef_r0.pt'
# model.load_state_dict(torch.load(state_dict_path, map_location=device))
model = torch.load(model_path, map_location=device)

model.eval()

y_true_list, y_pred_list = [], []

with torch.no_grad():
    for batch in testdata:
        # Handle both multi- and uni-modal batches
        if isinstance(batch, (list, tuple)):
            *mods, labels = batch
            # If it's your 3-modality baseline: (vision, audio, text, labels)
            # Move tensors to device if they are tensors
            mods_on_device = []
            for m in mods:
                if torch.is_tensor(m):
                    mods_on_device.append(m.to(device))
                else:
                    # If a modality is None or placeholder, keep it as is
                    mods_on_device.append(m)
            labels = labels.to(device) if torch.is_tensor(labels) else labels

            # Forward
            output = model(mods_on_device)
        else:
            raise RuntimeError("Unexpected batch format from testdata.")

        # Ensure 1D numpy arrays
        preds = output.detach().cpu().numpy().squeeze()
        trues = labels.detach().cpu().numpy().squeeze()

        # If model outputs shape [B,1], squeeze again
        if preds.ndim > 1 and preds.shape[-1] == 1:
            preds = preds[..., 0]
        if trues.ndim > 1 and trues.shape[-1] == 1:
            trues = trues[..., 0]


        y_pred_list.append(preds)
        y_true_list.append(trues)

y_true = np.concatenate(y_true_list)
y_pred = np.concatenate(y_pred_list)

print("Collected predictions:", y_true.shape, y_pred.shape,
      f"| mean true={y_true.mean():.3f}, mean pred={y_pred.mean():.3f}")
```

This code outlines the  procedure for evaluating a pre-trained multimodal model on a test dataset. The process begins with the critical step of re-establishing the exact architectural components—the encoders, fusion module, and head network—that were used during the model's training phase. This ensures that the newly instantiated model is structurally identical to the one whose parameters were saved. Subsequently, the script loads the entire pre-trained model object from a specified file path onto the appropriate computational device. Once loaded, the model is set to evaluation mode (.eval()), which is crucial for disabling layers with different training and inference behaviors, such as dropout, thereby guaranteeing consistent predictions. The core of the script is an inference loop that iterates through the test data batches, processing each input through the model without calculating gradients, a practice enabled by torch.no_grad() that significantly conserves memory and accelerates computation. As predictions are generated, they are meticulously collected alongside their ground-truth labels, with careful handling of data formats to ensure they are properly converted to NumPy arrays and flattened for subsequent metric calculations. The final output provides a summary of the collected data shapes and mean values, serving as an essential validation step before proceeding with a full performance analysis of the model's test-time performance.

```python
class DummyEncoder(nn.Module):
    def forward(self, x):
        return x.new_zeros((x.shape[0], 0)).cuda()

encoders = [
    DummyEncoder().cuda(),    # Vision(ignored)
    DummyEncoder().cuda(),    # Audio(ignored)
    nn.Sequential(
        GRU(300, 256, dropout=True, has_padding=False, batch_first=True, last_only=True),
        Identity()
    ).cuda()  # Text
]

fusion = Concat().cuda()
head = nn.Sequential(
    nn.Linear(256, 128),
    nn.ReLU(),
    nn.Dropout(0.47),
    nn.Linear(128, 1)
).cuda()

# Check output shape with one batch
for batch in traindata:
    features = encoders[2](batch[2].cuda())# this batch 2 means text
    output = head(features)
    print("Model output shape:", output.shape)  # Should print [batch_size, 1]
    print("Label shape:", batch[3].shape)       # Should print [batch_size, 1]
    break

train(encoders, fusion, head, traindata, validdata, 50,
      task="regression",                                  #Trains for 50 epochs on a regression task (continuous sentiment scores).
      optimtype=torch.optim.AdamW,                         #Optimizer: AdamW,Loss function: L1 Loss.aves the best model to text_gru.pt
      is_packed=False,
      lr=3e-4,
      save='text_gru.pt',
      weight_decay=0.01,
      objective=nn.L1Loss())

model = torch.load('text_gru.pt').cuda()
test(model, testdata, 'affect', is_packed=False, criterion=nn.L1Loss(),
     task="posneg-classification", no_robust=True)
```

This code trains a neural network that only "listens" to the text part of the data, ignoring the vision and audio. It does this by using "dummy" encoders that pass no information for those two parts. The core of the model is a GRU network that processes the text and then makes a prediction. The script trains this text-only model to predict a continuous score (like a sentiment rating). After training, it saves the best version of this model. The final step is to test how well this saved model works on a new dataset, specifically checking if its continuous scores can be used to correctly classify the data into simple categories like "positive" or "negative." This whole process is a way of seeing how important the text information is on its own, without any help from the other data types.

```python
# --- 3-Fusion: Vision + Audio + Text ---
#Takes a list of encoders (one each for vision, audio, text).
#Applies them separately to each modality input.
#Concatenates the extracted features (fusion).
#Passes the fused features to the final prediction layer (head).

#or each batch in traindata:
#Passes video (v), audio (a), and text (t) through the model.
#Computes MAE loss between predictions and ground truth labels (y).
#Updates model weights with backpropagation.
#Records training loss.

#Disables gradient updates (torch.no_grad()).
#Evaluates model on validation data.
#Tracks validation loss for each epoch.

import torch, torch.nn as nn
import numpy as np

class MultimodalModel(nn.Module):
    def __init__(self, encoders, fusion, head):
        super().__init__()
        self.encoders = nn.ModuleList(encoders)
        self.fusion = fusion
        self.head = head

    def forward(self, x_list):
        # x_list = [vision, audio, text]
        feats = []
        for i, enc in enumerate(self.encoders):
            xi = x_list[i]
            feats.append(enc(xi))
        z = self.fusion(feats)
        return self.head(z)  # expect [B,1]

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = MultimodalModel(encoders, fusion, head).to(device)

criterion = nn.L1Loss()
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-3, weight_decay=1e-5)

train_losses, val_losses = [], []
best_val = float("inf")
best_path = "fusion_vat_best.pt"

for epoch in range(1, 101):
    # ---- TRAIN ----
    model.train()
    running = 0.0
    for batch in traindata:
        v, a, t, y = batch
        v = v.to(device) if torch.is_tensor(v) else v
        a = a.to(device) if torch.is_tensor(a) else a
        t = t.to(device) if torch.is_tensor(t) else t
        y = y.to(device)

        optimizer.zero_grad()
        out = model([v, a, t])              # [B,1] ideally
        if out.ndim == 2 and out.shape[1] != 1:
            raise RuntimeError(f"Head returned {tuple(out.shape)}; expected [B,1]. Add a final nn.Linear(...,1).")
        loss = criterion(out, y)
        loss.backward()
        optimizer.step()
        running += float(loss.item())
    tr_loss = running / max(1, len(traindata))
    train_losses.append(tr_loss)

    # ---- VALIDATE ----
    model.eval()
    running = 0.0
    with torch.no_grad():
        for batch in validdata:
            v, a, t, y = batch
            v = v.to(device) if torch.is_tensor(v) else v
            a = a.to(device) if torch.is_tensor(a) else a
            t = t.to(device) if torch.is_tensor(t) else t
            y = y.to(device)
            out = model([v, a, t])
            loss = criterion(out, y)
            running += float(loss.item())
    va_loss = running / max(1, len(validdata))
    val_losses.append(va_loss)

    print(f"Epoch {epoch:03d} | train {tr_loss:.4f} | val {va_loss:.4f}")

    # save best by val loss
    if va_loss < best_val:
        best_val = va_loss
        torch.save(model.state_dict(), best_path)
        print("  ↳ Saved best checkpoint")

print(f"Best val MAE: {best_val:.4f}  (saved to {best_path})")
#This script provides a clean and standard training pipeline for a multimodal deep learning model.
# The MultimodalModel is a simple wrapper that orchestrates the three main stages of the forward pass: a list of encoders processes each raw data modality (e.g., vision, audio, text) into a compact feature vector;
# a fusion module combines these vectors into a single, unified representation; and a final head network uses this fused vector to make a prediction.
#The training process itself follows a well-established pattern. Over a series of epochs, the model is repeatedly trained on batches of data, with gradients calculated via nn.L1Loss and weights updated using the AdamW optimizer.
# Crucially, after each training epoch, the model's performance is evaluated on a separate validation dataset. This allows the script to implement an effective early-stopping strategy:
#it continuously tracks the lowest validation loss achieved so far and saves a checkpoint of the model's state only when a new best score is reached.
# This practice is essential for preventing the model from overfitting to the training data, ensuring that the final saved model is the one with the best generalization performance on unseen examples.
# The process concludes by reporting the best validation loss and the file path of the saved, optimal model, providing a complete and verifiable training recor
```

This script provides a clean and standard training pipeline for a multimodal deep learning model. The MultimodalModel is a simple wrapper that orchestrates the three main stages of the forward pass: a list of encoders processes each raw data modality (e.g., vision, audio, text) into a compact feature vector; a fusion module combines these vectors into a single, unified representation; and a final head network uses this fused vector to make a prediction. The training process itself follows a well-established pattern. Over a series of epochs, the model is repeatedly trained on batches of data, with gradients calculated via nn.L1Loss and weights updated using the AdamW optimizer. Crucially, after each training epoch, the model's performance is evaluated on a separate validation dataset. This allows the script to implement an effective early-stopping strategy: it continuously tracks the lowest validation loss achieved so far and saves a checkpoint of the model's state only when a new best score is reached. This practice is essential for preventing the model from overfitting to the training data, ensuring that the final saved model is the one with the best generalization performance on unseen examples. The process concludes by reporting the best validation loss and the file path of the saved, optimal model, providing a complete and verifiable training record.

##Attention based model

This code implements the attention-based multimodal fusion model for emotion recognition. Each modality (vision, audio, and text) is first processed by a GRU encoder, which transforms sequential input features into compact embeddings while handling variable sequence lengths through masked mean pooling. These embeddings are then passed into the ModalityAttentionFusion layer, which learns attention weights to highlight the relative importance of each modality for the task. The fused representation is fed into a multi-layer head network that produces the final prediction. The training loop supports both regression (predicting continuous sentiment scores) and binary classification (positive/negative sentiment), using appropriate loss functions (MAE, MSE, or BCE). It also incorporates AdamW optimisation, dropout regularisation, gradient clipping, and mixed-precision training to improve stability and efficiency. The evaluation function computes MAE, MSE, Pearson correlation, accuracy, weighted F1, and confusion matrices, enabling a comprehensive performance comparison. Overall, this code demonstrates how attention mechanisms can dynamically weight modalities rather than treating them equally, providing a more flexible alternative to the early fusion baseline.

```python


import math, numpy as np
from typing import Tuple, Dict, List
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
from torch.cuda.amp import autocast, GradScaler

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


# -------------------------
# Utils
# -------------------------
def _to_3d(x: torch.Tensor) -> Tuple[torch.Tensor, torch.Tensor]:
    """Accept [B,D] or [B,T,D]; return ([B,T,D], mask[B,T]) where mask=True means valid."""
    if x.dim() == 2:
        B, D = x.shape
        return x.unsqueeze(1), torch.ones(B, 1, dtype=torch.bool, device=x.device)
    if x.dim() == 3:
        mask = (x.abs().sum(dim=-1) > 0)
        return x, mask
    raise ValueError(f"Unexpected shape: {tuple(x.shape)}")

def masked_mean(x: torch.Tensor, mask: torch.Tensor, dim: int) -> torch.Tensor:
    w = mask.float().unsqueeze(-1)
    s = (x * w).sum(dim=dim)
    d = w.sum(dim=dim).clamp_min(1e-6)
    return s / d

@torch.no_grad()
def pearsonr_torch(x: torch.Tensor, y: torch.Tensor) -> float:
    x = x.float(); y = y.float()
    x = x - x.mean(); y = y - y.mean()
    denom = (x.norm() * y.norm()).clamp_min(1e-8)
    return float((x * y).sum() / denom)


# -------------------------
# Encoders
# -------------------------
class GRUEncoder(nn.Module):
    """[B,D] or [B,T,D] -> [B,H] via masked GRU + masked mean pooling."""
    def __init__(self, in_dim: int, hid: int = 128, num_layers: int = 1, dropout: float = 0.1):
        super().__init__()
        self.inp = nn.Linear(in_dim, hid)
        self.gru = nn.GRU(hid, hid, num_layers=num_layers, batch_first=True,
                          dropout=0.0 if num_layers == 1 else dropout, bidirectional=False)
        self.out = nn.Sequential(nn.LayerNorm(hid), nn.ReLU(), nn.Dropout(dropout), nn.Linear(hid, hid))
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        x3d, mask = _to_3d(x)
        h = self.inp(x3d) * mask.unsqueeze(-1).float()
        h, _ = self.gru(h)
        pooled = masked_mean(h, mask, dim=1)
        return self.out(pooled)


# -------------------------
# Attention Fusion
# -------------------------
class ModalityAttentionFusion(nn.Module):
    """Attention over modality embeddings (vision, audio, text). Returns fused vector + weights."""
    def __init__(self, hid: int):
        super().__init__()
        self.w = nn.Linear(hid, hid)
        self.v = nn.Linear(hid, 1, bias=False)
        self.bias = nn.Parameter(torch.zeros(hid))
        self.drop = nn.Dropout(0.1)
    def forward(self, mlist: List[torch.Tensor]):
        M = torch.stack(mlist, dim=1)         # [B,3,H]
        H = torch.tanh(self.w(M) + self.bias) # [B,3,H]
        scores = self.v(H).squeeze(-1)        # [B,3]
        alpha = torch.softmax(scores, dim=1)  # [B,3]
        fused = (M * alpha.unsqueeze(-1)).sum(dim=1)
        return self.drop(fused), alpha


# -------------------------
# Full Model
# -------------------------
class AttnFusionModel(nn.Module):
    def __init__(self, v_in: int, a_in: int, t_in: int, hid: int = 128, task: str = "regression"):
        super().__init__()
        assert task in ("regression", "binary")
        self.task = task
        self.ve = GRUEncoder(v_in, hid)
        self.ae = GRUEncoder(a_in, hid)
        self.te = GRUEncoder(t_in, hid)
        self.fusion = ModalityAttentionFusion(hid)
        self.head = nn.Sequential(
            nn.LayerNorm(hid), nn.ReLU(), nn.Dropout(0.2),
            nn.Linear(hid, hid), nn.ReLU(), nn.Dropout(0.2),
            nn.Linear(hid, 1)  # (B,1) then squeeze -> (B,)
        )
    def forward(self, v, a, t):
        ev, ea, et = self.ve(v), self.ae(a), self.te(t)  # each [B,H]
        fused, alpha = self.fusion([ev, ea, et])         # [B,H], [B,3]
        y = self.head(fused).squeeze(-1)                # [B]
        return y, alpha


# -------------------------
# Evaluation (full metrics)
# -------------------------
def _f1_weighted_np(y_true, y_pred):
    f1s, weights = [], []
    for cls in (0, 1):
        tp = np.sum((y_true==cls) & (y_pred==cls))
        fp = np.sum((y_true!=cls) & (y_pred==cls))
        fn = np.sum((y_true==cls) & (y_pred!=cls))
        prec = tp/(tp+fp) if (tp+fp)>0 else 0.0
        rec  = tp/(tp+fn) if (tp+fn)>0 else 0.0
        f1 = 2*prec*rec/(prec+rec) if (prec+rec)>0 else 0.0
        support = np.sum(y_true==cls)
        f1s.append(f1); weights.append(support)
    weights = np.array(weights, float)
    return float(np.average(np.array(f1s), weights=weights)) if weights.sum()>0 else 0.0

@torch.no_grad()
def evaluate(model, loader, task="regression", threshold=0.0) -> Dict[str, float]:
    model.eval()
    preds, golds = [], []
    for v, a, t, l in loader:
        v, a, t = v.to(device), a.to(device), t.to(device)
        l = l.to(device).float()
        yhat, _ = model(v, a, t)            # (B,)
        preds.append(yhat.detach().cpu())
        golds.append(l.view(-1).detach().cpu())  # force (B,)
    preds = torch.cat(preds); golds = torch.cat(golds)

    out = {
        "MAE": F.l1_loss(preds, golds).item(),
        "MSE": F.mse_loss(preds, golds).item(),
        "Pearson_r": pearsonr_torch(preds, golds)
    }

    if task == "binary":
        y_true = (golds.numpy() >= threshold).astype(np.int64)
        y_pred = (preds.numpy() >= threshold).astype(np.int64)
        acc = float((y_true == y_pred).mean())
        f1w = _f1_weighted_np(y_true, y_pred)
        # confusion matrix [[TN FP],[FN TP]]
        tn = int(((y_true==0) & (y_pred==0)).sum())
        fp = int(((y_true==0) & (y_pred==1)).sum())
        fn = int(((y_true==1) & (y_pred==0)).sum())
        tp = int(((y_true==1) & (y_pred==1)).sum())
        out.update({"Accuracy": acc, "F1_weighted": f1w, "Confusion_Matrix": [[tn, fp],[fn, tp]]})
    return out


# -------------------------
# Training
# -------------------------
def train_attn_fusion(
    traindata: DataLoader, validdata: DataLoader, testdata: DataLoader,
    v_in: int, a_in: int, t_in: int,
    task: str = "regression",          # "regression" or "binary"
    loss_type: str = "mae",            # "mae", "mse", or "bce" (binary)
    epochs: int = 20, lr: float = 2e-3, hid: int = 128, wd: float = 1e-4,
    threshold: float = 0.0, use_amp: bool = True,
):
    model = AttnFusionModel(v_in, a_in, t_in, hid=hid, task=task).to(device)

    # Criterion
    if task == "binary" or loss_type == "bce":
        criterion = nn.BCEWithLogitsLoss(); use_bce = True
    elif loss_type == "mse":
        criterion = nn.MSELoss(); use_bce = False
    else:
        criterion = nn.L1Loss(); use_bce = False

    opt = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=wd)
    scaler = GradScaler(enabled=(use_amp and device.type == "cuda"))

    best_val = None
    best_state = None
    train_hist, val_hist = [], []

    def _prep_labels(l: torch.Tensor, yshape: torch.Size):
        l = l.float().view(-1)  # (B,)
        if use_bce:
            l = (l >= threshold).float()
        assert l.shape == yshape, f"label {tuple(l.shape)} != pred {tuple(yshape)}"
        return l

    for ep in range(1, epochs+1):
        # train
        model.train()
        loss_sum, n = 0.0, 0
        for v, a, t, l in traindata:
            v, a, t, l = v.to(device), a.to(device), t.to(device), l.to(device)
            opt.zero_grad(set_to_none=True)
            with autocast(enabled=(use_amp and device.type=="cuda")):
                yhat, _ = model(v, a, t)           # (B,)
                lprep = _prep_labels(l, yhat.shape)
                loss = criterion(yhat, lprep)
            scaler.scale(loss).backward()
            nn.utils.clip_grad_norm_(model.parameters(), 5.0)
            scaler.step(opt); scaler.update()
            bs = yhat.size(0)
            loss_sum += loss.item() * bs; n += bs
        tr_loss = loss_sum / max(n,1)
        train_hist.append(tr_loss)

        # val (criterion-aligned)
        model.eval()
        v_sum, vn = 0.0, 0
        with torch.no_grad():
            for v, a, t, l in validdata:
                v, a, t, l = v.to(device), a.to(device), t.to(device), l.to(device)
                yhat, _ = model(v, a, t)
                lprep = _prep_labels(l, yhat.shape)
                vloss = criterion(yhat, lprep)
                v_sum += vloss.item() * yhat.size(0); vn += yhat.size(0)
        va_loss = v_sum / max(vn,1)
        val_hist.append(va_loss)

        val_metrics = evaluate(model, validdata, task=task, threshold=threshold)
        msg = (f"[Epoch {ep:02d}] train_loss={tr_loss:.4f} | val_loss={va_loss:.4f} | "
               f"Val MAE={val_metrics['MAE']:.4f} | MSE={val_metrics['MSE']:.4f} | r={val_metrics['Pearson_r']:.3f}")
        if task == "binary":
            msg += f" | Acc={val_metrics['Accuracy']:.3f} | F1w={val_metrics['F1_weighted']:.3f}"
        print(msg)

        # keep best by val_loss
        score = -va_loss
        if best_val is None or score > best_val:
            best_val = score
            best_state = {k: v.detach().cpu() for k,v in model.state_dict().items()}

    if best_state is not None:
        model.load_state_dict({k: v.to(device) for k,v in best_state.items()})

    print("\nBest model (by val_loss) evaluation:")
    val_metrics = evaluate(model, validdata, task=task, threshold=threshold)
    test_metrics = evaluate(model, testdata, task=task, threshold=threshold)
    print("  VALID:", val_metrics)
    print("  TEST :", test_metrics)

    history = {"train_loss": train_hist, "val_loss": val_hist}
    return model, {"val": val_metrics, "test": test_metrics}, history


# -------------------------
# Helpers
# -------------------------
def infer_input_dims(loader: DataLoader):
    """Peek one batch to infer (Dv, Da, Dt). Loader must yield (vision, audio, text, labels)."""
    v,a,t,_ = next(iter(loader))
    def d(x): return x.shape[-1] if x.dim() in (2,3) else (_ for _ in ()).throw(ValueError(f"Bad shape {x.shape}"))
    return d(v), d(a), d(t)
```

This code implements a robust deep learning model designed to fuse information from multiple modalities: vision, audio, and text. The system's architecture can be broken down into three key stages: encoding, attention-based fusion, and final prediction.

First, the encoding stage uses dedicated `GRUEncoder` modules for each modality. These encoders process time-series data (like a sequence of frames for video or audio features) and condense them into a single, fixed-length vector. This is analogous to a feature extractor, transforming raw, high-dimensional data into a more manageable, abstract representation.

Next, the core of the model is the attention-based fusion mechanism. This is where the magic happens. The model takes the three encoded vectors (one each for vision, audio, and text) and calculates a dynamic "attention score" for each. Essentially, the model learns to weigh the importance of each modality for a specific task and input. For example, if the model is predicting sentiment, and the text is neutral while the audio has a very strong emotional tone, the attention mechanism will give the audio features more weight. It then creates a single, fused vector by combining the three modality vectors based on these learned weights.

Finally, the prediction stage takes this fused vector and passes it through a small neural network to produce the final output. This output can be a continuous value for a regression task (e.g., predicting a numerical score) or a binary output for a classification task (e.g., predicting a positive or negative label).

The code also includes a comprehensive training and evaluation pipeline. It uses advanced techniques like Automatic Mixed Precision (AMP) to speed up training, gradient clipping to prevent instability, and an early-stopping mechanism to ensure the model doesn't overfit the training data. The `evaluate` function provides a full suite of standard metrics (MAE, MSE, Pearson's r, F1-weighted score) to thoroughly assess the model's performance on unseen validation and test data, providing a complete and verifiable solution.

##Results

```python
import numpy as np
import torch

@torch.no_grad()
def eval_with_confusion(model, loader, threshold=0.0, device="cuda"):
    """
    Evaluate model with Accuracy, F1_weighted, MAE, CorrCoeff, and Confusion Matrix.
    """
    model.eval()
    preds, golds = [], []

    for batch in loader:
        v, a, t, l = batch
        v, a, t, l = v.to(device), a.to(device), t.to(device), l.to(device).float()
        yhat, _ = model(v, a, t)
        preds.append(yhat.view(-1).cpu())
        golds.append(l.view(-1).cpu())

    preds = torch.cat(preds)
    golds = torch.cat(golds)

    # ---- Regression metrics ----
    mae = torch.nn.functional.l1_loss(preds, golds).item()
    r = torch.corrcoef(torch.stack([preds, golds]))[0,1].item()

    # ---- Binary metrics ----
    y_true = (golds.numpy() >= threshold).astype(int)
    y_pred = (preds.numpy() >= threshold).astype(int)

    acc = (y_true == y_pred).mean()

    # Weighted F1
    f1s, weights = [], []
    for cls in [0, 1]:
        tp = np.sum((y_true==cls) & (y_pred==cls))
        fp = np.sum((y_true!=cls) & (y_pred==cls))
        fn = np.sum((y_true==cls) & (y_pred!=cls))
        prec = tp/(tp+fp) if (tp+fp)>0 else 0
        rec  = tp/(tp+fn) if (tp+fn)>0 else 0
        f1   = 2*prec*rec/(prec+rec) if (prec+rec)>0 else 0
        support = np.sum(y_true==cls)
        f1s.append(f1); weights.append(support)
    f1_weighted = np.average(f1s, weights=weights)

    # ---- Confusion Matrix ----
    tn = int(((y_true==0) & (y_pred==0)).sum())
    fp = int(((y_true==0) & (y_pred==1)).sum())
    fn = int(((y_true==1) & (y_pred==0)).sum())
    tp = int(((y_true==1) & (y_pred==1)).sum())
    cm = [[tn, fp],[fn, tp]]

    # Normalized confusion matrix (rows sum to 1)
    cm_norm = [[cm[0][0]/max((cm[0][0]+cm[0][1]),1), cm[0][1]/max((cm[0][0]+cm[0][1]),1)],
               [cm[1][0]/max((cm[1][0]+cm[1][1]),1), cm[1][1]/max((cm[1][0]+cm[1][1]),1)]]

    return {
        "MAE": mae,
        "CorrCoeff": r,
        "Accuracy": acc,
        "F1_weighted": f1_weighted,
        "Confusion_Matrix": cm,
        "Confusion_Matrix_Normalized": cm_norm
    }
```

```python
results_val = eval_with_confusion(model, validdata, threshold=0.0)
results_test = eval_with_confusion(model, testdata, threshold=0.0)

print("VALID:", results_val)
print("TEST :", results_test)
```
